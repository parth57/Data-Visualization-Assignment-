```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X_red = PCA(3).fit_transform(iris['data'])
y = iris['target']
```

```python
axes = Axes3D(plt.figure(figsize = (12,6)))
axes.scatter(X_red[:,0], X_red[:,1], X_red[:,2], c = y, cmap = 'rainbow')
axes.set_title('PCA')
axes.set_xlabel('First Dimension')
axes.set_ylabel('Second Dimension')
axes.set_zlabel('Third Dimension')
```
